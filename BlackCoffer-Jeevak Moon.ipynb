```python
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd

# read the input excel file
input_df = pd.read_excel('input.xlsx')

# loop through each row of the input file
for index, row in input_df.iterrows():

    # get the URL and URL_ID from the current row
    url = row['URL']
    url_id = row['URL_ID']

    # send a request to the URL and get the response
    response = requests.get(url)

    # parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # find the main content of the article and exclude the header and footer
    article_title = soup.find('h1').text.strip()
    article_text = ''
    article_content = soup.find('div', {'class': 'td-post-content'})
    # modify the class name as per your HTML structure
    for paragraph in article_content.find_all('p'):
        article_text += paragraph.text.strip() + '\n'

    # create a directory to save the extracted article text
    if not os.path.exists('articles'):
        os.makedirs('articles')

    # save the extracted article text to a text file with URL_ID as its file name
    with open(f'articles/{url_id}.txt', 'w', encoding='utf-8') as f:
        f.write(article_title + '\n\n' + article_text)
```

# Now extracting text from the given URLs and saving it in another folder

```python
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd

# read the input excel file
input_df = pd.read_excel('input.xlsx')
input_df_1 = input_df.drop(7)
input_df_2 = input_df_1.drop(20)
input_df_3 = input_df_2.drop(107)

# loop through each row of the input file
for index, row in input_df_3.iterrows():

    # get the URL and URL_ID from the current row
    url = row['URL']
    url_id = row['URL_ID']

    # send a request to the URL and get the response
    response = requests.get(url)

    # parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # find the main content of the article and exclude the header and footer
    article_title = soup.find('h1').text.strip()
    article_text = ''
    article_content = soup.find('div', {'class': 'td-post-content'})
    # modify the class name as per your HTML structure
    for paragraph in article_content.find_all('p'):
        article_text += paragraph.text.strip() + '\n'

    # create a directory to save the extracted article text
    if not os.path.exists('articles'):
        os.makedirs('articles')

    # save the extracted article text to a text file with URL_ID as its file name
    with open(f'articles/{url_id}.txt', 'w', encoding='utf-8') as f:
        f.write(article_title + '\n\n' + article_text)
```

# Now performing text analysis on the extracted text 

```python
import os
import glob
import string
import zipfile
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
from openpyxl import load_workbook
from textstat.textstat import textstatistics

# Load the positive and negative words
positive_words = set()
negative_words = set()
with open('extracted_files/MasterDictionary/positive-words.txt', 'r', encoding='ANSI') as pos_file:
    positive_words.update(set(pos_file.read().splitlines()))
with open('extracted_files/MasterDictionary/negative-words.txt', 'r', encoding='ANSI') as neg_file:
    negative_words.update(set(neg_file.read().splitlines()))

# Load the stopwords
stop_words = set(stopwords.words('english'))

# Load the existing workbook and select the active worksheet
wb = load_workbook(filename='Output Data Structure.xlsx')
ws = wb.active

# Set up Textstat
ts = textstatistics()

# Clean the text and remove punctuation and stopwords
def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = word_tokenize(text)
    words = [word for word in words if word not in stop_words]
    cleaned_text = ' '.join(words)
    return cleaned_text
            
# Function to calculate derived variables
def calculate_derived_variables(text):
    blob = TextBlob(text)
    sentences = sent_tokenize(text)
    words = word_tokenize(text)
    
    positive_word_count = sum(1 for word in words if word.lower() in positive_words)
    negative_word_count = sum(1 for word in words if word.lower() in negative_words)
    polarity_score = (positive_word_count - negative_word_count) / (positive_word_count + negative_word_count + 0.000001)
    subjectivity_score = (positive_word_count + negative_word_count) / (len(words) + 0.000001)
    return polarity_score, subjectivity_score

# Function to calculate positive score
def calculate_positive_score(tokens, positive_words):
    return sum(1 for word in tokens if word in positive_words)
            
# Function to calculate negative score
def calculate_negative_score(tokens, negative_words):
    return sum(1 for word in tokens if word in negative_words)

# Function to calculate average sentence length
def calculate_average_sentence_length(text):
    sentences = sent_tokenize(text)
    num_sentences = len(sentences)
    num_words = len(word_tokenize(text))
    return num_words / num_sentences

# Function to calculate the percentage of complex words
def calculate_percentage_complex_words(text):
    words = word_tokenize(text)
    complex_word_count = sum(1 for word in words if ts.syllable_count(word) > 2)
    return (complex_word_count / len(words)) * 100, complex_word_count

# Function to calculate Fog Index
def calculate_fog_index(avg_sentence_length, percentage_complex_words):
    return 0.4 * (avg_sentence_length + percentage_complex_words)

# Function to calculate average number of words per sentence
def calculate_average_words_per_sentence(text):
    sentences = sent_tokenize(text)
    num_sentences = len(sentences)
    num_words = len(word_tokenize(text))
    return num_words / num_sentences

# Function to count total words
def count_total_words(text):
    words = word_tokenize(text)
    return len(words)

# Function to count personal pronouns
def count_personal_pronouns(text):
    words = word_tokenize(text)
    personal_pronouns = ['i', 'me', 'my', 'mine', 'we', 'us', 'our', 'ours']
    num_personal_pronouns = sum(1 for word in words if word.lower() in personal_pronouns)
    return num_personal_pronouns

# Function to calculate average word length
def calculate_average_word_length(text):
    words = word_tokenize(text)
    num_words = len(words)
    total_word_length = sum(len(word) for word in words)
    return total_word_length / num_words

# Function to calculate syllables per word
def calculate_syllables_per_word(text):
    words = word_tokenize(text)
    num_syllables = sum(ts.syllable_count(word) for word in words)
    num_words = len(words)
    return num_syllables / num_words

# Loop through all text files in the folder
for i, filename in enumerate(glob.glob('articles/*.txt')):
    with open(filename, 'r', encoding='utf-8') as file:
        text = file.read() 
        
    # Tokenize the cleaned text
    tokens = word_tokenize(cleaned_text)    
    cleaned_text = clean_text(text)
    positive_score = calculate_positive_score(tokens, positive_words)
    negative_score = calculate_negative_score(tokens, negative_words)
    polarity_score, subjectivity_score = calculate_derived_variables(cleaned_text)
    avg_sentence_length = calculate_average_sentence_length(cleaned_text)
    percentage_complex_words, complex_word_count = calculate_percentage_complex_words(cleaned_text)
    fog_index = calculate_fog_index(avg_sentence_length, percentage_complex_words)
    avg_words_per_sentence = calculate_average_words_per_sentence(cleaned_text)
    word_count = count_total_words(cleaned_text)
    personal_pronouns = count_personal_pronouns(cleaned_text)
    avg_word_length = calculate_average_word_length(cleaned_text)
    syllable_per_word = calculate_syllables_per_word(cleaned_text)
    
    # Save the results to the XLSX file
    row = i + 2  # Adjust row index to leave row 1 empty
    column = 3  # Start from column 3
    ws.cell(row=row, column=column).value = positive_score
    ws.cell(row=row, column=column + 1).value = negative_score
    ws.cell(row=row, column=column + 2).value = polarity_score
    ws.cell(row=row, column=column + 3).value = subjectivity_score
    ws.cell(row=row, column=column + 4).value = avg_sentence_length
    ws.cell(row=row, column=column + 5).value = percentage_complex_words
    ws.cell(row=row, column=column + 6).value = fog_index
    ws.cell(row=row, column=column + 7).value = avg_words_per_sentence
    ws.cell(row=row, column=column + 8).value = complex_word_count
    ws.cell(row=row, column=column + 9).value = word_count
    ws.cell(row=row, column=column + 10).value = syllable_per_word
    ws.cell(row=row, column=column + 11).value = personal_pronouns
    ws.cell(row=row, column=column + 12).value = avg_word_length

# Save the modified workbook
wb.save(filename='Output Data Structure.xlsx')
```
